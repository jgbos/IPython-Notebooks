# Do Field Goal Percentages Drop For Cold Weather Games?

This is a variation of the analysis done by [Greg Reda's](http://www.gregreda.com/2013/12/26/three-pointers-after-offensive-rebounds/) 3-point analysis for NCAA Basketball. Lets get started:

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import pandas.io.sql as psql
import mysql.connector as sql

%matplotlib inline
import matplotlib.pyplot as plt

# from mpld3 import enable_notebook
# enable_notebook()
```

Yes, I paid the money for NFL data set from [Armchair Analysis](http://armchairanalysis.com/).  The nice thing about their data is that it is broken up nicely for doing quick analysis, and they provide scripts for setting up and importing their data into a SQL database.  The following code loads the data from MySQL to a pandas DataFrame object using pandas.io.sql.

```python
import pickle
with open('sql.config','rb') as fp:
    config = pickle.load(fp)

con = sql.connect(**config)
df = psql.frame_query('''
    SELECT fgxp.DIST, fgxp.GOOD, games.TEMP FROM fgxp, core, games 
    WHERE core.PID=fgxp.PID AND games.GID = core.GID AND fgxp.FGXP='FG';''', con)

con.close()
```

The data for temperature was loaded into the SQL database as character string, convert it to a number.

```python
df.TEMP = df.TEMP.astype(float)
df.head()
```

## MCMC Simulation

Now lets partition the data for doing a MCMC study.  Lets look at distances less than 40 yards and define cold temperatures as those less than 40 degrees.

```python
convert = lambda x: True if x == 'Y' else False

normal = (df.TEMP >= 40) & (df.DIST < 40) & df.TEMP.notnull()
normal = df[normal].GOOD.apply(convert)

cold = (df.TEMP < 40) & (df.DIST < 40) & df.TEMP.notnull()
cold = df[cold].GOOD.apply(convert)

print normal.mean(), normal.std()
print cold.mean(), cold.std()
```

It looks like there is a difference in the mean, but these are the observations of a [Bernoulli Trial](http://en.wikipedia.org/wiki/Bernoulli_distribution). Lets do a MCMC simulation to estimate the probability of making a field goal in normal versus cold conditions.

```python
import pymc as pm
```

```python
with pm.Model() as model:
    p_normal = pm.Uniform("p_normal",lower=0.7, upper=1)
    p_cold = pm.Uniform("p_cold",lower=0.7, upper=1)
    
    # scraped observations
    obs_normal = pm.Bernoulli("obs_normal", p_normal, observed=normal.astype(int))
    obs_cold = pm.Bernoulli("obs_cold", p_cold, observed=cold.astype(int))
    
    m = pm.HamiltonianMC()
    trace = pm.sample(15000, m,progressbar=False)
```

```python
p_normal_samples = trace['p_normal'][:]
p_cold_samples=  trace['p_cold'][:]
delta_samples = trace['p_normal']-trace['p_cold']
```

```python
plt.figure(figsize=(12.5, 10))

ax = plt.subplot(311)
plt.xlim(0.8, 1)
plt.xticks(np.arange(0.8, 1.0, 0.01))
plt.ylim(0, 200)
plt.hist(p_normal_samples, histtype='stepfilled', bins=50, normed=True, 
         color='#E41A1C', label='Warm')
plt.vlines(normal.mean(), 0, 300, linestyles='--')
plt.legend()
plt.grid(False)

ax = plt.subplot(312)
plt.xlim(0.8, 1)
plt.xticks(np.arange(0.8, 1.0, 0.01))
plt.ylim(0, 200)
plt.hist(p_cold_samples, histtype='stepfilled', bins=50, normed=True,
         color='#4DAF4A', label='Cold')
plt.vlines(cold.mean(), 0, 300, linestyles='--')
plt.legend()
plt.grid(False)

ax = plt.subplot(313)
plt.xlim(-0.1, 0.1)
plt.xticks(np.arange(-0.1, 0.101, 0.01))
plt.ylim(0, 100)
plt.hist(delta_samples, histtype='stepfilled', bins=50, normed=True,
         color='#377EB8', label='Delta')
plt.vlines(0, 0, 300, linestyles='--', label='$H_0$ (No difference)')
plt.legend()
plt.grid(False);
```

```python
print ("Field Goals less than 40 yards are less successful in temperatures < 40 degrees "
       "in {0:.1f}% of simulations").format((delta_samples > 0).mean() * 100)
```

It appears that there is significant difference in the probability of successful field goals in cold temperatures, a roughly 3% difference.

## Using Classification

Now lets look at using an ensemble learning method, Random Forests, to estimate the probability of successfully making a field goal.  This method will be able to train a model given temperature and distance.  Additionally, as Trey Causey points out on [the spread](www.thespread.us), Random Forests can give you a confidence interval for your prediction.  

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
convert = lambda x: 1 if x == 'Y' else 0
data = df;
data = data[data.TEMP.notnull()]
data.TEMP = data.TEMP.astype(float)
data.GOOD = data.GOOD.apply(convert)

features = data[['DIST','TEMP']]
target = data['GOOD']

# Set up the random forest model with parameters found via grid search
model = RandomForestClassifier(oob_score = True, verbose = 1,  criterion = 'entropy', n_estimators=100, n_jobs=2, min_samples_leaf=100)
model = model.fit(features, target)

model.score(features, target)
```

```python
p = np.array([])
p_low = np.array([])
p_high = np.array([])
p1 = np.array([])
p1_low = np.array([])
p1_high = np.array([])
s = np.array([])
for i in range(20,60,5):
    s = np.hstack([s,i])
    n = []
    for t in range(0,40,5):
        for estimator in model.estimators_:
            n.append(estimator.predict_proba([i,t])[0][1])
    n = np.array(n)    
    p_low = np.hstack([p_low,np.percentile(n,2.5)])
    p_high = np.hstack([p_high,np.percentile(n,97.5)])        
    p = np.hstack([p,n.mean()])
    
    n = []
    for t in range(40,110,5):
        for estimator in model.estimators_:
            n.append(estimator.predict_proba([i,t])[0][1])
    n = np.array(n)    
    p1_low = np.hstack([p1_low,np.percentile(n,2.5)])
    p1_high = np.hstack([p1_high,np.percentile(n,97.5)])        
    p1 = np.hstack([p1,n.mean()])

plt.figure(figsize=(8, 6))
plt.plot(s,p, label="Cold")
plt.fill_between(s, p_low, p_high, alpha=0.2, label="Cold Uncertainty")
plt.plot(s,p1, label="Warm")
plt.fill_between(s, p1_low, p1_high, alpha=0.2, label="Warm Uncertainty", color="green")
plt.ylim(0.3,1)
plt.grid()
plt.xlabel('FG Distance')
plt.ylabel('Probability of Making FG')
plt.legend()
```

This is an interesting result, the mean values are close to that estimated in this [talk](http://www.sloansportsconference.com/wp-content/uploads/2013/Slides/RP/Going%20for%20Three_Predicting%20the%20likelihood%20of%20field%20goal%20success%20with%20logistic%20regression.pdf).  The uncertainty region here doesn't quite make sense, the warm temperatures have a larger uncertainty but have a lot more data.  Also, they do not match the MCMC simulation uncertainty.  I will have to look into a better aproach to estimate the confidence interval with the Random Forest model.
