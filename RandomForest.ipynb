# Using Random Forests to estimate probability of making a 3-point shot after an offensive rebound

This is a variation of the analysis done by [Greg Reda's](http://www.gregreda.com/2013/12/26/three-pointers-after-offensive-rebounds/) 3-point analysis for NCAA Basketball.  Greg used a MCMC method to analyize the probability of making a 3-point shot after an offensive rebound.  Here I attempt to use the Random Forest using [scikit-lear](http://scikit-learn.org/dev/modules/ensemble.html).

Lets get started:

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import glob
import re

%matplotlib inline
import matplotlib.pyplot as plt

from mpld3 import enable_notebook
enable_notebook()
```

Load web scraped data using Greg's script.

```python
# read PSVs into DataFrame
games = []
files = glob.glob('cbb-play-data/*.psv')
for f in files:
    df = pd.read_csv(f, sep='|')
    df['game_id'] = f.replace('.psv', '')
    games.append(df)

print 'Read {0} games'.format(len(games))
```

```python
games_df = pd.concat(games)

# add event_id to maintain event order
# we can use the index since pandas defaults to the Nth row of the file
games_df['event_id'] = games_df.index

# melt data into one column for home/away and another for event
# maintain play order by sorting on event_id
melted = pd.melt(games_df, id_vars=['event_id', 'game_id', 'time', 'score'],
                     var_name='team', value_name='event')
melted.sort_index(by=['game_id', 'event_id'], inplace=True)

# drop rows with NaN events - an event only belongs to one team
melted = melted[melted.event.notnull()]
print melted[10:15]
```

Now label the data whether a 3-point shot was made or missed.

```python
def c(r):
    if r=="made":
        return 1
    else:
        return 0
    
# label whether three pointers were made or missed
get_shot_result = lambda x: c(re.findall('(made|missed)', x)[0])
shot3 = melted.event.str.contains('Three Point')
melted['shot_result'] = melted[shot3].event.apply(get_shot_result)
```

Now label whether the shot was made after an offensive rebound.

```python
def criteria(df):
    """Labels if the three pointer was preceded by an offensive rebound."""
    df['after_oreb'] = ((df.event.str.contains('Three Point')) & \
                        df.event.shift(1).str.contains('Offensive Rebound')) + 0
    df.after_oreb.fillna(0, inplace=1)
    return df

melted = melted.groupby('game_id').apply(criteria)
melted[melted.shot_result.notnull()].head(3)
```

Now lets take a look at the data:

```python
threes = melted[melted.shot_result.notnull()]
attempts = threes.groupby(['shot_result', 'after_oreb']).size().unstack(0)

made = attempts.columns[1]
missed = attempts.columns[0]
attempts['perc'] = attempts[made].astype(float) / (attempts[missed] + attempts[made])
print attempts
```

This shows that normally (not after an offensive rebound) the percentage of shots made is 34.8%, and 33.9% for after an offensive rebound.  Not much difference in percentages.

One thing to note, the data does not account for how long it took to make a 3-point shot after an offensive rebound.  We only care about those shots that are "immediately" after the rebound.  To do this we must calculate the time difference between each play.  

```python
melted['minutes'] = melted.time.apply(lambda x: int(x.split(':')[0]))
melted['seconds'] = melted.time.apply(lambda x: int(x.split(':')[1]))
```

There are also duplicates in the data, remove them.

```python
duped_cols = ['game_id', 'event_id', 'time', 'event']
melted[melted.duplicated(cols=duped_cols)][:3]
```

```python
melted.drop_duplicates(cols=['game_id', 'event_id', 'event'], inplace=True)
```

Label which period the play was made in order to calculate the total sime remaining.

```python
melted['period_end'] = melted.event.apply(lambda x: x.startswith('End of'))
melted[melted.period_end].head(3)
```

```python
calculate_period = lambda x: x.shift(1).cumsum().fillna(0) + 1
melted['period'] = melted.groupby('game_id').period_end.apply(calculate_period)
melted[melted.period_end].head(3)
```

```python
calculate_period = lambda x: x.shift(1).cumsum().fillna(0) + 1
melted['period'] = melted.groupby('game_id').period_end.apply(calculate_period)
```

```python
melted.set_index('game_id', inplace=True)

# 40min regulation game + (# periods - 2 halves) * 5min OTs
gametime = lambda x: 40 + (x - 2) * 5
melted['gametime'] = melted.groupby(level=0).period.max().apply(gametime)
melted.reset_index('game_id', inplace=True)
melted.groupby('gametime').game_id.nunique()
```

There are issues with the play-by-play data as descirbed by Greg, this fixes those issues.

```python
melted.loc[melted.gametime == 35, 'gametime'] = 40
```

Now calculate the time left in seconds in each period.

```python
def clock_to_secs_left(df):
    """Calculates the total seconds left in the game."""
    df['secs_left'] = np.nan
    df.loc[df.period == 1, 'secs_left'] = (df.minutes * 60) + 1200 + df.seconds
    df.loc[df.period > 1, 'secs_left'] = (df.minutes * 60) + df.seconds
    return df
```

```python
clock_to_secs_left(melted)
print melted[['game_id', 'time', 'event', 'period', 'secs_left']][:5]
```

Now we can calculate the time elapsed since the last play.  Label all 3-point shots attempted within 7 seconds of a rebound.

```python
melted['secs_elapsed'] = melted.secs_left.shift(1) - melted.secs_left
```

```python
mask = (melted.secs_elapsed >= 0) & (melted.secs_elapsed <= 7)
melted['recent'] = mask.astype(int)
data = melted[melted.shot_result.notnull()]
```

This is where we deviate from Greg's analysis.  Instead of using pyMC to calculate the distribution (and probably the best option for his analysis), lets try running the Random Forest classifier.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
features = data[['after_oreb','recent']]
target = data['shot_result']

# Set up the random forest model with parameters found via grid search
model = RandomForestClassifier(oob_score = True, verbose = 1,  criterion = 'entropy', n_estimators=100, n_jobs=2, min_samples_leaf=100)
model = model.fit(features, target)

model.score(features, target)
```

Now lets calclulate the probability of making a 3-pointer after an offensive rebound and normally.

```python
probs_normal = []
probs_after = []
for estimator in model.estimators_:
    probs_normal.append(estimator.predict_proba([0,0])[0][1])
    probs_after.append(estimator.predict_proba([1,1])[0][1])
```

```python
x = np.linspace(0.3,0.4,200)
mean_normal = np.array(probs_normal).mean()
std_normal = np.array(probs_normal).std()
mean_after = np.array(probs_after).mean()
std_after = np.array(probs_after).std()

plt.figure(figsize=(12.5, 10))

ax = plt.subplot(311)
plt.xlim(0.3, 0.4)
plt.xticks(np.arange(0.3, 0.401, 0.01))
plt.plot(x,stats.norm.pdf(x,mean_normal, std_normal))
plt.grid(False)

ax = plt.subplot(312)
plt.xlim(0.3, 0.4)
plt.xticks(np.arange(0.3, 0.401, 0.01))
plt.plot(x,stats.norm.pdf(x, mean_after, std_after))
plt.grid(False)

x = np.linspace(-0.05,0.05,200)
ax = plt.subplot(313)
plt.xlim(-0.05, 0.05)
plt.xticks(np.arange(-0.05, 0.0501, 0.01))
plt.plot(x,stats.norm.pdf(x,mean_after-mean_normal, (std_normal**2 + std_after**2)**0.5))
plt.grid(False)
```

```python
print ("3P% after offensive rebounds is more successful "
       "about {0:.1f}% of the time").format((1-stats.norm.cdf(0,mean_after-mean_normal,(std_normal**2 + std_after**2)**0.5)) * 100)
```

The results are very similar to Gregs histogram, very promising.

Now lets change the analysis a bit, lets play around with our classifier.  What if we add the time remaining as a feature so we can see if there are certain times in a game that shooting a 3-pointer is more successful.

```python
features1 = data[['after_oreb','recent', 'secs_left']]
target1 = data['shot_result']

# Set up the random forest model with parameters found via grid search
# model1 = RandomForestClassifier(n_estimators=100)
model = RandomForestClassifier(oob_score = True, verbose = 1,  criterion = 'entropy', n_estimators=100, n_jobs=2, min_samples_leaf=100)
model1 = model.fit(features1, target1)

model1.score(features1, target1)
```

```python
pn = np.array([])
pa = np.array([])
pd = np.array([])
pe = np.array([])
s = np.array([])
for i in range(0,2400,120):
    s = np.hstack([s,i])
    a = []
    n = []
    for estimator in model1.estimators_:
        n.append(estimator.predict_proba([0,0,i])[0][1])
        a.append(estimator.predict_proba([1,1,i])[0][1])
    n = np.array(n)
    a = np.array(a)    
    
    pn = np.hstack([pn,n.mean()])
    pa = np.hstack([pa, a.mean()])
    pd = np.hstack([pd, a.mean()-n.mean()])
    pe = np.hstack([pe, 3*(a.std()**2 + n.std()**2)**0.5])

plt.figure(figsize=(12.5, 10))
plt.plot(s,pd, label="Probability")
plt.fill_between(s, pd-pe, pd+pe,alpha=0.2, label="Uncertainty")
plt.ylim(-1,1)
plt.vlines(20*60, -1, 1, linestyles='--', label='Halftime')
plt.xlabel('Time Remaining (s)')
plt.ylabel('Added Probability of Taking 3-PT Shot')
plt.legend()
```

First, I clearly do not have enough data (1 season) to make any real determination.  The uncertainty is just too large.
